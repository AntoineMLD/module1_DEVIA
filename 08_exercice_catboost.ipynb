## Contexte Professionnel :

Vous êtes sollicité par une usine de compote de pommes aux États-Unis qui souhaite optimiser son processus de production en améliorant la gestion de la qualité des pommes qu'elle reçoit des agriculteurs. L'entreprise envisage d'automatiser le processus de classification et d'évaluation des pommes pour les utiliser dans différentes gammes de compotes, notamment des compotes premium pour les pommes de qualité supérieure et des compotes premiers prix pour d'autres variétés.

## Problématique :

Le processus actuel de classification des pommes est manuel, ce qui peut entraîner des incohérences et des erreurs humaines. L'usine souhaite mettre en place des modèles de machine learning pour automatiser cette classification et prédire la qualité des pommes en fonction de caractéristiques telles que la taille, le poids, la douceur, la croustillance, la jutosité, la maturité, l'acidité, etc.


## Tâche :

En tant que responsable de projet, vous utiliserez un ensemble de données fourni par l'usine, comprenant des informations détaillées sur diverses variétés de pommes. Les caractéristiques incluent la taille, le poids, la douceur, la croustillance, la jutosité, la maturité, l'acidité et la qualité. Vous développerez des modèles de classification et de prédiction de qualité basés sur ces données pour contribuer à l'amélioration des pratiques de l'usine.

## Consigne :

Pour relever ce défi, vous serez amené(e) à utiliser deux approches distinctes pour la modélisation : CatBoost, un puissant algorithme de boosting, et une Régression Logistique, un modèle classique mais souvent efficace.

## Tâches à Réaliser :

1. **Préparation des Données :** Assurez-vous que les données sont correctement chargées et nettoyées. Convertissez les variables catégorielles en formats adaptés si nécessaire.

2. **Modélisation avec CatBoost :** Utilisez l'algorithme CatBoost pour développer un modèle de classification des pommes basé sur les différentes caractéristiques fournies dans le jeu de données. Expérimentez avec les hyperparamètres pour optimiser les performances du modèle. Analyser les features qui ont le plus de poids dans la prédiction (get_feature_importance)

3. **Modélisation avec Régression Logistique :** Implémentez une Régression Logistique pour la classification des pommes. Effectuez le prétraitement des données requis et ajustez le modèle.

4. **Évaluation des Modèles :** Comparez les performances des deux modèles en termes de précision. Utilisez les métriques appropriées pour évaluer la qualité des classifications réalisées.
